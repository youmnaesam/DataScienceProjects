```python
import pandas as pd
import re

def clean_title_and_description(row):
    title = row['Title']
    description = row['Description']
    authors = row['Authors']

    # Convert NaN â†’ empty string
    title = "" if pd.isna(title) else str(title)
    description = "" if pd.isna(description) else str(description)
    authors = "" if pd.isna(authors) else str(authors)

    # 1. Split title on first ':'
    if ":" in title:
        parts = title.split(":", 1)
        title = parts[0].strip()
        subtitle = parts[1].strip()

        # Add subtitle to description
        if description:
            description = description + " " + subtitle
        else:
            description = subtitle

    # 2. Remove author names from title
    # BUT: Don't remove if it's possessive (e.g., "Amy Spangler's Breastfeeding")
    if authors.strip():
        # Check if title starts with possessive form (Author's Something)
        # Pattern: starts with name + 's + word
        possessive_pattern = r"^[\w\s]+[''']s\s+\w+"

        if not re.match(possessive_pattern, title):
            # Only remove author names if NOT possessive form
            author_list = re.split(r'[:;,&/]', authors)

            for author in author_list:
                author = author.strip()
                if author and len(author) > 2:  # Avoid removing single letters
                    # Remove author name with word boundaries
                    pattern = r'\b' + re.escape(author) + r'\b'
                    title = re.sub(pattern, "", title, flags=re.IGNORECASE)

    # Clean extra spaces and punctuation artifacts
    title = " ".join(title.split()).strip()

    # Remove leading/trailing punctuation
    title = re.sub(r'^[\'"\s\-:,]+|[\'"\s\-:,]+$', '', title)

    return pd.Series([title, description])

# Apply the function
df[['Clean_Title', 'Clean_Description']] = df.apply(clean_title_and_description, axis=1)

# Show results
print("Results:")
print(df[['Title', 'Authors', 'Clean_Title', 'Clean_Description']].head(10))
```

```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
tqdm.pandas()

# Robust predict function
def predict_category(author_name):
    if not isinstance(author_name, str) or author_name.strip() == '':
        return np.nan  # return NaN for invalid input
    
    # remove common prefix "By " and strip spaces
    author_name_clean = author_name.replace("By ", "").strip()
    
    # transform single author safely (sparse)
    vec = vectorizer.transform([author_name_clean])
    
    # predict
    pred_id = model.predict(vec)[0]
    pred_label = le.inverse_transform([pred_id])[0]
    
    return pred_label

# Select rows with missing categories
missing_mask = df['Category'].isna()
missing_authors = df.loc[missing_mask, 'Authors']

# Predict categories for missing authors
predicted_categories = missing_authors.progress_apply(predict_category)

# Create a new dataframe with results
predicted_df = pd.DataFrame({
    'Authors': missing_authors,
    'Predicted_Category': predicted_categories
})

print(predicted_df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend(book_idx, top_n=5):
    # Compute similarity between ONE book and all others
    sims = cosine_similarity(
        embeddings[book_idx].reshape(1, -1),
        embeddings
    )[0]
    
    # Get top N indices
    top_indices = np.argsort(sims)[::-1][1:top_n+1]
    
    return df.iloc[top_indices]['Title'].tolist()

recommend(10, top_n=5)
```

```python
from sklearn.preprocessing import normalize
import numpy as np

embeddings = normalize(embeddings)

def recommend_by_title(book_title, top_n=5):
    # Case-insensitive search
    matches = df[df['Title'].str.lower().str.contains(book_title.lower())]
    
    if matches.empty:
        return "Book not found."
    
    # Take first match
    book_idx = matches.index[0]
    
    # Compute similarity
    sims = embeddings @ embeddings[book_idx]
    
    # Get top similar books (exclude itself)
    top_indices = np.argsort(sims)[::-1][1:top_n+1]
    
    return df.iloc[top_indices]['Title'].tolist()

recommend_by_title("harry potter", top_n=5)
```

```python
from difflib import get_close_matches
import numpy as np

def recommend_by_title(book_title, top_n=5, cutoff=0.6):
    # Create lowercase title list
    title_list = df['Title'].str.lower().tolist()
    
    # Find closest match
    matches = get_close_matches(book_title.lower(), title_list, n=1, cutoff=cutoff)
    
    if not matches:
        return "Book not found. Try a more accurate title."
    
    # Get the index of the matched book
    book_idx = df[df['Title'].str.lower() == matches[0]].index[0]
    
    # Compute similarity
    sims = embeddings @ embeddings[book_idx]
    top_indices = np.argsort(sims)[::-1][1:top_n+1]
    
    recommendations = df.iloc[top_indices]['Title'].tolist()
    
    print(f"\nSelected Book: {df.loc[book_idx, 'Title']}")
    print("\nRecommended Books:")
    for i, title in enumerate(recommendations, 1):
        print(f"{i}. {title}")
    
    return recommendations

recommend_by_title("The Sorcerer's Stone")
```
